```python
M = []
I = 1
for it in range(I):
    for antigen in Ag:
        d = dist(antigen, Ab)
        f = affinity (d)
        Cn = int(sum(1/f))
        Abn = np.argsort(-f)[:n]
        fsum = sum(f[Abn])
        fdiv = (f[Abn]/fsum) * Cn
        left = np.random.randint(3)
        C = []
        C.append(Ab[Abn[left]])
        for i in  range(0, n):
            J = int(fdiv[i])
            for j in range(J):
                Ck = Ab[Abn[i]]
                C.append (Ck + (1/f[Abn[i]]) * (antigen - Ck))
        C = np.array(C)
        d = affinity (dist(antigen, C))
        m = C[np.argsort(-d)[:math.ceil(len(C) * qi)]]
        d = memAffinity (m)
        exc = exclusion(d)
        m = [m[i] for i in range (len(m)) if i not in exc]
        M += m
    d = memAffinity (M)
    exc = exclusion(d)
    Ab = M
    M = [M[i] for i in range (len(M)) if i not in exc]

    
```
